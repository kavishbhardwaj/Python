Key Points to Remember:
Short Form: Typically a single character preceded by a hyphen (-), e.g., -l, -a, -R.

Long Form: A descriptive word preceded by two hyphens (--), e.g., --all, --recursive, --long.

Short and long forms can be used interchangeably. For example, -l is equivalent to --long.

Multiple short options can be combined (e.g., ls -laR), but long options must be written separately (e.g., ls --long --all --recursive).

Use man ls or ls --help to see all available short and long forms for ls.

Short Form vs. Long Form of ls Options

Short Form	Long Form	Description
-a	          --all	    Show all files, including hidden files (those starting with .).

-1	--format=single-column	List one file per line.

This sequence includes a broad range of options for ls:

Explanation of ls with Directories as Arguments:


ls <directory_name>: Lists the contents of the specified directory, not the current working directory.
Example: ls level1 lists files inside level1 directory.


ls <dir1> <dir2>: Lists contents of multiple directories.
Example: ls level1 subdir1 lists files in both level1 and subdir1.


ls -ld <directory>: Lists information about the directory itself, not its contents.
Example: ls -ld level1 shows details (like permissions, owner, etc.) of the directory level1 itself.


Recursive Listing (ls -R <directory>): Displays all files and subdirectories within the specified directory recursively.
Example: ls -R level1 shows all files in level1 and its subdirectories.


Combine Options with Directory Arguments:
ls -alhR level1 subdir1: Shows all hidden files in a long format, human-readable sizes, and recursively lists contents of level1 and subdir1.


#### Structure of Each Line given by ls -l:
Structure will change slightly depending on VS code is run in wsl or not

1. **Permissions**: (`drwxr-xr-x`)
   - This shows the file permissions for the owner, group, and others.
   - first letter -  `d` indicates a **directory** (if it were a file, it would be `-`).
                     'l' indicates symbolic link (shortcut)
                     '-' indicates file 
                     'c' character device (terminal)
                     'b' block devices (hard discs)
   - next three for owner - `rwx` means **read, write, and execute** permissions for the owner.
   - next three for group - `r-x` means **read and execute** permissions for the group.
   - next three for others- `r-x` means **read and execute** permissions for others.

   rwx (7)
   rw- (6)
   r-x (5)
   r-- (4)
   --x (1)

2. **Links**: (`1`)
   - The number of links or directories inside this directory. For a directory, this number (1) represents how many subdirectories, parent directories, or other references link to it. A hard link is like an alias or shortcut that points to the same file or directory in the filesystem. 

3. **Owner**: (`kavis`)
   - The username of the owner of the file or directory.

4. **Group**: (`197609`)
   - The group ID associated with the file.

5. **Size**: (`0`, `40`, `12288`, etc.)
   - The size of the file in bytes. For directories, itâ€™s usually `0` since it doesnâ€™t directly reflect the size of contents.

6. **Timestamp**: (`Oct 6 14:21`)
   - The date and time when the file or directory was last modified.

7. **File/Directory Name**: (`$Recycle.Bin`, `Documents and Settings`, etc.)
   - The name of the file or directory.

#### Listing Explanation:

**`Documents and Settings -> /c/Users`**:
   - This is a **symbolic link** (shortcut) that points to `/c/Users`. It means that trying to access `Documents and Settings` will actually redirect you to the `/c/Users` directory.

**`Drivers`**:
   - Likely a folder containing drivers for the system. Drivers are software that help the operating system communicate with hardware devices.

**`Program Files`** and **`Program Files (x86)`**:
   - Standard directories in Windows for installing 64-bit and 32-bit applications, respectively.

**`ProgramData`**:
   - A directory used to store application data that is not specific to a user. Itâ€™s shared among all users of the system.

**`Users`**:
    - This is where user profiles are stored on Windows, such as `Desktop`, `Documents`, `Downloads`, and so on for each user.

**`Windows`**:
    - The main directory containing the Windows operating system files.

**`hiberfil.sys`**:
    - A **system file** used by Windows to store the system state when it goes into **hibernation**. This file is often large and contains everything stored in memory (RAM).

### inode and hardlink
The inode tracks the file or directory, while the hard link count tells you how many different references or paths point to that inode.

### inode
Each inode number (e.g., 12, 8193, 8194) uniquely identifies a file or directory on the filesystem, like a unique ID.

ls -i shows the inode number associated with each file or directory, which helps identify them uniquely


### hard link
Each directory starts with at least 2 hard links:

One for the directory itself (.)
One for its parent directory (..)
Every subdirectory inside it adds 1 more hard link to the count. This is because each subdirectory contains a link (..) pointing back to the parent directory. 

Example:
var has 13 hard links.
It includes . and ...
And 11 subdirectories inside var, each adding 1 to the count.

When you use ls -l, the number of hard links is shown in the second column. For example, in drwxr-xr-x 13 root root 4096 Oct 5 12:49 var, the 13 means there are 13 hard links to the var directory. The inode for var is linked to every file or directory that shares it.

### Difference Between Linux and Windows File Systems

### File System Types:

Linux: Uses file systems like ext4, ext3, xfs, and btrfs.
Windows: Uses NTFS (New Technology File System) and FAT32.


### Permissions
Linux: Has a detailed permission model with read, write, and execute permissions for the user, group, and others. It supports commands like chmod and chown to control access.
Windows: Uses Access Control Lists (ACLs) to set permissions. ACLs allow for more granular control but are not compatible with the Unix permission model. This is why changing permissions with chmod on a Windows file system has no effect.


### Case Sensitivity:

Linux: Case-sensitive. file.txt and File.txt are treated as separate files.
Windows: Generally case-insensitive. file.txt and File.txt are considered the same file.

### File Path Differences:

Linux: Paths use forward slashes (/), e.g., /home/kavis.
Windows: Paths use backslashes (\), e.g., C:\Users\kavis.

When you work in directories like /mnt/c/, /mnt/d/, etc., youâ€™re actually interacting with the Windows file system through WSL, which has limitations when it comes to Unix-style operations.
By working in Linux directories like /home/kavis, you can use native Linux commands and features (like chmod) without restrictions.
In Summary: Move to directories like /home/kavis to use Linux file system features effectively and see your chmod changes reflect properly!

```python
%%bash

cd
pwd


# Create a test directory structure with nested subdirectories and files
mkdir -p test_dir/level1/level2
cd test_dir
touch file1 file2 level1/file3 level1/level2/file4
mkdir subdir1 subdir2

# 1. Default `ls` with no options
ls

# 2. Using `-l -t` vs `-t -l`:
#    `-l` is for long format and `-t` is for sorting by modification time.
#    These two options can be used in any order and the output will be the same.
ls -lt  # Long format sorted by modification time
ls -tl  # Sorted by modification time with long format

# 3. Using `-l` and `-S` in different orders:
#    `-S` sorts by size and `-l` shows long format. The order does not matter here.
ls -lS  # Long format sorted by file size
ls -Sl  # Sorted by file size in long format

# 4. Combining options `-l`, `-a`, and `-t` in different orders:
#    All options will work in any order, and the output will be the same.
ls -lat  # Long format, including hidden files, sorted by modification time
ls -tal  # Sorted by modification time, long format, including hidden files
ls -alt  # Including hidden files, long format, sorted by modification time

# 5. Using `-d` with directory arguments:
#    `-d` changes how directories are listed. The order of `-d` matters if you include it with other options.
#    When `-d` is specified after other options, it will change how the directories themselves are displayed.
ls -ld level1 subdir1  # Show details of directories level1 and subdir1, not their contents
ls -dl level1 subdir1  # Same output as above since -d modifies the behavior of -l

# 6. Recursive listing (`-R`) combined with other options in different orders:
#    `-R` will recursively list all contents. Combining with other options like `-l`, `-a`, or `-S` can be in any order.
ls -lRa  # Long format, recursive, including hidden files
ls -aRl  # Including hidden files, recursive, long format

# Clean up: remove the created test directory
cd ..
rm -r test_dir
```

Key Takeaways on Option Ordering:
Most options can be combined in any order:

For example, ls -lS, ls -Sl, and ls -l -S will produce the same output because -l and -S do not conflict.
Some options modify the behavior of other options:

ls -d (show directory itself) when combined with -l (ls -ld) will show details of the directory itself rather than its contents. But using -l -d or -d -l does not change the behavior.
Grouping of options:

Options can be grouped together as ls -l -t or ls -lt. There is no difference in output when options are grouped or not, as long as the options do not modify each otherâ€™s behavior.
Effect of -R (recursive):

The order of -R does not matter when combined with other options like -l or -a. All contents will be listed recursively no matter where -R appears.
Options that change display behavior (e.g., -d, -l):

These options need to be understood carefully, as -d will override the behavior of other options like -R (no recursion).
Summary
The order of options does not matter in most cases, but options like -d can change how others behave, so be mindful when combining.

Explanation of Recursive Listing Options:


ls -R: This is the basic recursive listing. It displays all files and directories under the current directory and its subdirectories. Itâ€™s useful for getting a complete overview of the entire directory tree.


ls -lR: Combines the long format listing (-l) with recursive listing. This provides a detailed view of permissions, owners, sizes, and modification timestamps for all files and directories recursively.


ls -aR: Shows all files, including hidden ones (files starting with .), recursively.


ls -laR: Combines long format and hidden file listing with recursive listing.


ls -R <directory_name>: Lists the contents of a specific directory (e.g., dir1) recursively. It will show all files and subdirectories under dir1.


ls -lhR: Combines long format with human-readable file sizes (KB, MB, etc.) for all files and directories recursively.


ls -ltR: Combines sorting by modification time (-t) with recursive listing. This will show the most recently modified files first within each directory.


ls -lrR: Combines reverse sorting (-r) with recursive listing. This will display files in reverse alphabetical order within each directory.


ls -dR: Using -d with -R will only display the directory names themselves, not their contents. This option can be useful if you want to list the directories without listing everything inside.


ls -sR: Shows file sizes in blocks, recursively. This is useful for getting an idea of storage consumption in each directory.


When to Use Recursive Listing:
To see the structure of nested directories.
To analyze large directory trees for storage usage or permissions.
To find specific files deep within a directory structure without manually traversing each directory.
Using recursive listing helps in scenarios where you want to understand the overall structure and contents of a directory tree.

```python
%%bash
# Create a sample directory structure for demonstration
cd
mkdir -p arg_demo/subdir1 arg_demo/subdir2
cd arg_demo

# Create sample files for testing
echo "This is file1" > file1.txt
echo "This is file2" > file2.txt
echo "This is file3" > file3.txt

# 1. Multiple Arguments
# Using `mv` and `cp` commands with multiple source files as arguments
ls                # List files to see what's present

# Move multiple files (file1.txt and file2.txt) to subdir1
mv file1.txt file2.txt subdir1/
ls subdir1        # Check that both files were moved to subdir1

# Copy multiple files (file1.txt and file2.txt) back to the main directory
cp subdir1/file1.txt subdir1/file2.txt .
ls                # Check that both files were copied back to the main directory

# 2. Second Argument
# The second argument is typically the target location for `mv` and `cp` commands.
# Move a file to subdir2, where subdir2 is the second argument.
mv file3.txt subdir2/
ls subdir2        # Check that file3.txt was moved to subdir2

# Copy the file back to the main directory, using subdir2/file3.txt as source and . as the second argument.
cp subdir2/file3.txt .
ls                # Check that file3.txt was copied back to the main directory

# 3. Interpretation of Last Argument
# The last argument is interpreted as the target directory if multiple sources are provided.
# Move file1.txt and file2.txt into subdir2 using subdir2/ as the last argument.
mv file1.txt file2.txt subdir2/
ls subdir2        # Check that both files are moved to subdir2

# Copy file1.txt and file2.txt back to the main directory using . as the last argument.
cp subdir2/file1.txt subdir2/file2.txt .
ls                # Check that both files are copied back to the main directory

# 4. Recursion Assumed for `mv` but Not for `cp`
# Create nested directories and files for demonstration of recursion
mkdir -p subdir1/nested_dir
echo "Nested file content" > subdir1/nested_dir/nested_file.txt

# Moving a directory with contents is straightforward with `mv`:
mv subdir1/nested_dir subdir2/
ls subdir2/nested_dir  # Check that nested_dir and its contents were moved to subdir2

# Copying a directory requires the `-r` or `-R` flag for recursion:
# Copy nested_dir from subdir2 back to subdir1 using `-r` for recursion.
cp -r subdir2/nested_dir subdir1/
ls subdir1/nested_dir  # Check that nested_dir and its contents were copied back to subdir1

# Clean up: remove the created arg_demo directory
cd ..
rm -r arg_demo
```

Explanation of Each Section:
Multiple Arguments:

The mv and cp commands can take multiple source files as arguments followed by a destination directory.
Example: mv file1.txt file2.txt subdir1/ moves both file1.txt and file2.txt to subdir1.
Second Argument:

The second argument is typically the destination or target for the operation.
Example: mv file3.txt subdir2/ moves file3.txt to the subdir2 directory.
Interpretation of Last Argument:

The last argument is always interpreted as the target if multiple sources are provided.
Example: mv file1.txt file2.txt subdir2/ moves file1.txt and file2.txt into subdir2 because the last argument (subdir2/) is considered the destination.
Recursion Assumed for mv but Not for cp:

mv assumes recursion when moving directories with contents. You can move a directory without specifying additional flags.
Example: mv subdir1/nested_dir subdir2/ moves the entire nested_dir with its contents to subdir2.
cp requires -r or -R flag for recursion. Without the -r flag, copying a directory will result in an error.
Example: cp -r subdir2/nested_dir subdir1/ copies the entire nested_dir with its contents back to subdir1.
Summary of Commands Used:
mv:

Moves files or directories from one location to another.
Recursively moves directories by default.
cp:

Copies files from one location to another.
Requires -r or -R for copying directories and their contents.
ls:

Lists the contents of directories to verify file operations.

```python
%%bash
# Create a sample directory structure for demonstration
cd
mkdir -p link_demo
cd link_demo

# Create a sample file for linking demonstration
echo "This is the original file." > original_file.txt
echo "This is another file." > another_file.txt

# 1. **Hard Links**:
# Create a hard link to the original file
# Syntax: ln <source_file> <hard_link_name>
ln original_file.txt hard_link.txt

# List files to see if hard_link.txt is created
ls -li            # Use -i to show inode numbers

# Display contents of both files to verify they are identical
cat original_file.txt
cat hard_link.txt

# Make a change to the hard link and verify if it affects the original file
echo "Appending text to hard link." >> hard_link.txt

# Display contents again to verify changes reflect in both files
cat original_file.txt
cat hard_link.txt

# Check inode numbers of both files to confirm they are the same
ls -li original_file.txt hard_link.txt

# 2. **Symbolic Links (Soft Links)**:
# Create a symbolic link to the original file
# Syntax: ln -s <source_file> <symbolic_link_name>
ln -s original_file.txt soft_link.txt

# List files to see if soft_link.txt is created
ls -l            # Use -l to show symbolic links with the arrow (->)

# Display contents of the symbolic link to verify it points to the original file
cat soft_link.txt

# Make a change to the original file and see if it reflects in the symbolic link
echo "Appending text to original file." >> original_file.txt

# Display contents again to verify changes reflect in the symbolic link
cat soft_link.txt

# Check inode numbers of all three files to confirm hard link shares inode but symbolic link does not
ls -li original_file.txt hard_link.txt soft_link.txt

# 3. **Delete the Original File**:
# Delete the original file and see what happens to the hard and symbolic links
rm original_file.txt

# List files to see the result
ls -l

# Display contents of the remaining files
cat hard_link.txt      # Should work as it shares the same inode as the original file
cat soft_link.txt      # Will show an error since the original file is deleted

# Clean up: remove the created link_demo directory
cd ..
rm -r link_demo
```

Explanation of Each Section:
Creating Hard Links:

Command: ln original_file.txt hard_link.txt
A hard link is another name for the same file. Both the original file and the hard link share the same inode number.
Changes to the hard link reflect in the original file and vice versa.
If you delete the original file, the hard link still exists and retains the file content.
Creating Symbolic (Soft) Links:

Command: ln -s original_file.txt soft_link.txt
A symbolic (soft) link is a pointer to the original file. It does not share the same inode number as the original file.
If you delete the original file, the symbolic link becomes a "broken link" and no longer works.
Symbolic links are displayed with an arrow (->) pointing to the target file in the output of ls -l.
Verifying Links:

Command: ls -li and cat <filename>
The -i option shows inode numbers, allowing you to verify that hard links share the same inode number while symbolic links do not.
Displaying file contents (cat) demonstrates that changes made to the hard link reflect in the original file, while the symbolic link depends on the presence of the original file.
Effect of Deleting the Original File:

After deleting the original file (rm original_file.txt):
Hard Link: The hard link continues to exist because it shares the inode with the original file.
Symbolic Link: The symbolic link becomes broken because it points to a deleted file.
Summary of Commands Used:
Creating Hard Links:

ln <source_file> <hard_link_name>: Creates a hard link named <hard_link_name> pointing to <source_file>.
Creating Symbolic Links:

ln -s <source_file> <symbolic_link_name>: Creates a symbolic link named <symbolic_link_name> pointing to <source_file>.
Verifying Links:

ls -li: Shows files with their inode numbers to confirm hard links share the same inode.
cat <file_name>: Displays file contents to show the relationship between links and the original file.
Deleting Original File:

rm <file_name>: Deletes the original file to see the behavior of hard and symbolic links.

```python
%%bash
# Create a sample directory structure for demonstration
cd
mkdir -p size_demo
cd size_demo

# Create sample files of varying sizes
echo "This is a small file." > small_file.txt
dd if=/dev/zero of=medium_file.txt bs=512 count=100  # Creates a 50 KB file
dd if=/dev/zero of=large_file.txt bs=1024 count=200  # Creates a 200 KB file

# List the created files
ls -lh            # List files with human-readable sizes

# 1. **File Size Using `ls -s`**:
# Display file sizes in blocks with `ls -s`
ls -s             # Shows the size of each file in 1KB blocks (default)

# Use `-h` to display file sizes in human-readable format along with `-s`
ls -sh            # Shows sizes in human-readable format

# List with detailed view including file sizes and other attributes
ls -lsh           # Combine -l, -s, and -h to show detailed information with file sizes

# 2. **File Size Using `stat`**:
# Display detailed information including file size and block size
stat small_file.txt   # Shows file size in bytes, allocated blocks, and block size
stat medium_file.txt  # Shows file size in bytes, allocated blocks, and block size
stat large_file.txt   # Shows file size in bytes, allocated blocks, and block size

# 3. **File Size Using `du` (Disk Usage)**:
# Display disk usage of each file in blocks
du small_file.txt     # Shows the disk usage of small_file.txt
du medium_file.txt    # Shows the disk usage of medium_file.txt
du large_file.txt     # Shows the disk usage of large_file.txt

# Show disk usage in human-readable format
du -h small_file.txt
du -h medium_file.txt
du -h large_file.txt

# Show total size of the current directory and its contents
du -sh .             # Total disk usage of the current directory

# 4. **Role of Block Size**:
# Create a new file with a specific block size using `dd`
dd if=/dev/zero of=block_size_file.txt bs=4096 count=1  # Creates a file with 1 block of 4KB

# Display block size using `stat` to see allocation
stat block_size_file.txt

# Compare `stat` and `du` output for the file with a single block
stat block_size_file.txt
du block_size_file.txt

# Check file size in blocks using `ls -s`
ls -s block_size_file.txt

# Clean up: remove the created size_demo directory
cd ..
rm -r size_demo
```

Explanation of Each Command:
File Size Using ls -s:

ls -s shows the size of each file in blocks. By default, each block is 1KB on most systems.
ls -sh displays file sizes in human-readable format (KB, MB, GB).
ls -lsh combines -l, -s, and -h to show detailed information along with file sizes in blocks and human-readable format.
File Size Using stat:

stat provides detailed information about a file, including:
File size: Actual file size in bytes.
Blocks allocated: Number of blocks allocated to store the file on disk.
IO Block size: The block size used for file system I/O operations.
This command helps you understand the difference between the file size and the space occupied due to block allocation.
File Size Using du (Disk Usage):

du shows the actual disk usage of a file, which may be larger than the file's size due to the file system's block size.
du -h <filename> displays the disk usage in a human-readable format.
du -sh . shows the total disk usage of the current directory and its contents.
Role of Block Size:

Creating a file with a specific block size using dd helps illustrate how block size affects disk usage.
stat displays the block size used for I/O operations and the number of blocks allocated.
du shows the actual disk usage, which will reflect the allocated blocks.
Key Concepts:
File Size vs. Disk Usage:

The actual file size is the number of bytes in the file.
Disk usage is the amount of disk space the file occupies, which depends on the file system's block size.
Block Size:

Files are stored in blocks on disk. If a file is smaller than the block size, it still occupies one full block.
For example, a 1-byte file on a system with a 4KB block size will occupy 4KB of disk space.
Understanding Block Size with stat and du:

stat shows file size, blocks allocated, and I/O block size.
du shows the actual disk usage in terms of blocks allocated.

```python
%%bash
# Exploring the /proc filesystem

# 1. List contents of the /proc directory to see various files and subdirectories
ls -l /proc

# 2. Display the contents of some common files in /proc
# These files provide system information, process status, and other kernel parameters.

# Display system information from the /proc/cpuinfo file
cat /proc/cpuinfo | head -n 20    # Show the first 20 lines of CPU information

# Display memory information from the /proc/meminfo file
cat /proc/meminfo | head -n 20    # Show the first 20 lines of memory information

# Show the current kernel version from the /proc/version file
cat /proc/version

# Display uptime information from the /proc/uptime file
cat /proc/uptime

# Display current processes and their status
# The /proc directory contains subdirectories named after each process ID (PID).
ls -l /proc | grep "^[0-9]"        # List only the directories representing PIDs

# Display detailed information of a specific process (e.g., process ID 1)
cat /proc/1/status                 # Show the status of process with PID 1

# Display the command line of a specific process (e.g., process ID 1)
cat /proc/1/cmdline

# 3. Inspecting files in /proc for additional system details
# Display partitions information from /proc/partitions
cat /proc/partitions

# Display current mounts from /proc/mounts
cat /proc/mounts

# 4. Exploring the /sys filesystem
# /sys is used to expose kernel modules, devices, and other hardware information.

# List contents of the /sys directory
ls -l /sys

# List contents of /sys/class to see different device classes
ls -l /sys/class

# Display information about a specific device (e.g., CPU)
ls -l /sys/devices/system/cpu

# Show all CPUs present on the system
ls -l /sys/devices/system/cpu/cpu*

# Display information about a specific CPU (e.g., CPU0)
cat /sys/devices/system/cpu/cpu0/topology/core_id
cat /sys/devices/system/cpu/cpu0/cpufreq/scaling_governor

# 5. Check block device information using /sys/block
ls -l /sys/block

# Show details about a specific block device (e.g., sda)
ls -l /sys/block/sda

# Clean up: No cleanup required since /proc and /sys are in-memory filesystems
```

Explanation of Each Command:
Exploring /proc Filesystem:

The /proc directory is a virtual filesystem that provides information about the kernel, system, and running processes. The contents are not actual files but rather kernel-generated information.
The following files and directories are explored:
/proc/cpuinfo: Shows detailed information about the CPUs on the system.
/proc/meminfo: Shows memory-related information such as total memory, free memory, and cached memory.
/proc/version: Displays the Linux kernel version.
/proc/uptime: Shows system uptime in seconds.
/proc/[PID]: Each subdirectory corresponds to a running process identified by its Process ID (PID).
/proc/partitions: Displays partition information.
/proc/mounts: Shows all mounted filesystems and their properties.
Exploring /sys Filesystem:

The /sys directory is another virtual filesystem used to provide information about devices, kernel modules, and hardware attributes.
Common subdirectories in /sys:
/sys/class: Represents different classes of devices, such as cpu, net, and block.
/sys/devices: Provides detailed information about devices connected to the system.
/sys/block: Represents block devices (e.g., storage devices).
Checking Hardware and Device Information:

/sys/devices/system/cpu/cpu*: Lists all CPUs present on the system.
/sys/block/sda: Represents a block device such as a hard disk or SSD.
Detailed Information on Processes:

Each process running on the system has a corresponding directory in /proc with its PID as the directory name.
status, cmdline, stat, and other files within these directories provide information about the process, such as its command-line arguments, memory usage, and more.
Key Concepts:
In-memory Filesystems:

/proc and /sys are virtual filesystems that exist in memory. They do not occupy disk space and are used by the kernel to expose information and configuration options.
Role of /proc:

/proc primarily provides information about the kernel, system resources, and running processes. It is read-only, and changing files here will not persist across reboots.
Role of /sys:

/sys provides a structured view of kernel modules, devices, and hardware attributes. It allows interaction with hardware settings (e.g., changing CPU frequency).
