## What are Flags and Arguements?

Flags and arguments are essential components of Linux commands.

*   **Flags** (or options) modify the behavior of a command. They usually start with a hyphen (`-`) and can be either single (`-l`) or double (`--long`).
*   **Arguments** are inputs that a command requires to execute. They usually follow the command and flags.



````python
# @title Example
%%bash
ls -l /home
````

In this example:

*   `ls` is the command.
*   `-l` is a flag that tells `ls` to use a long listing format.
*   `/home` is an argument specifying the directory to list.

## Argument - Parameter Extension

(Taken from the GNU Bash Manual) [Link](https://www.gnu.org/software/bash/manual/html_node/Shell-Parameter-Expansion.html)

The ‘$’ character introduces parameter expansion, command substitution, or arithmetic expansion. The parameter name or symbol to be expanded may be enclosed in braces, which are optional but serve to protect the variable to be expanded from characters immediately following it which could be interpreted as part of the name.

When braces are used, the matching ending brace is the first ‘}’ not escaped by a backslash or within a quoted string, and not within an embedded arithmetic expansion, command substitution, or parameter expansion.

The basic form of parameter expansion is ${parameter}. The value of parameter is substituted. The parameter is a shell parameter as described above or an array reference. The braces are required when parameter is a positional parameter with more than one digit, or when parameter is followed by a character that is not to be interpreted as part of its name.

Note (Refer to this part when variables are taught in the lectures. ) -

If the first character of parameter is an exclamation point (!), and parameter is not a nameref, it introduces a level of indirection. Bash uses the value formed by expanding the rest of parameter as the new parameter; this is then expanded and that value is used in the rest of the expansion, rather than the expansion of the original parameter. This is known as indirect expansion. The value is subject to tilde expansion, parameter expansion, command substitution, and arithmetic expansion. If parameter is a nameref, this expands to the name of the variable referenced by parameter instead of performing the complete indirect expansion. The exceptions to this are the expansions of `${!prefix*}` and `${!name[@]}` described below. The exclamation point must immediately follow the left brace in order to introduce indirection.

Arguments in the context of command-line interfaces and scripting refer to the values or inputs that you pass to a command or script when you execute it. These arguments allow you to provide additional information or specify options for the command. In bash, arguments are typically represented as positional parameters, which can be accessed using the `$1`, `$2`, etc., notation (we will revisit this concept more during the lectures), where `$1` corresponds to the first argument, `$2` to the second, and so on. For example, if you have a script named `greet.sh` that takes a name as an argument, you might run it like this:

```
bash greet.sh Alice
```
Inside the script, you could access the name using `$1`, allowing you to personalize the output, such as:

```
echo "Hello, $1!"
```
This would produce the output: `Hello, Alice!` Here, `Alice` is the argument passed to the script.

The bash variable will come in Week-5, for the sake of completeness, we have included it here. We will deal with this topic separately.

````python
# @title Example
%%bash
name="World"
echo "Hello, $name"
````

In this example, `$name` is replaced by the value of the variable `name`. Further, you can also use curly braces (more standard practice) to include the variable name.

````python
# @title Advanced Example
%%bash
touch file1.txt file2.txt file3.txt file4.pdf file5.pdf
files="*.txt" # More information on * (a wild character) will be discussed later.
echo $files
````

Here, `$files` will be expanded to match all `.txt` files in the current directory.

**Note** -

Commands can vary significantly in terms of the number and type of arguments they accept. Some commands require multiple arguments to function properly, while others may not need any at all. For instance, the `cp` command, which is used to copy files, typically requires at least two arguments: the source file and the destination. For example:

```
cp file.txt /path/to/destination/
```

In this case, `file.txt` is the source, and `/path/to/destination/` is the destination where the file will be copied. On the other hand, commands like `date`, which prints the current working directory, do not take any arguments:

```
date
```

Here, executing `date` without any arguments will simply return the date. Further, format can be specified using `+FORMAT`.

Therefore, understanding how many arguments a command requires or allows is crucial for effective command-line usage.

## Brace Expansion

Brace expansion is a mechanism by which sequences of characters are generated.

Syntax:



```
{start..end}
```



````python
# @title Example1
%%bash
echo {1..5}
````

````python
# @title Example2
%%bash
echo file{1..3}.txt
````


In bash, the `echo` command is used to display text or the value of variables in the terminal. When using braces for parameter expansion, you can perform operations or manipulate variables within the `echo` command. This expansion can even be nested, allowing for more complex expressions. For example, consider the following:

```
name="Alice"
echo "Hello, ${name^^}!"
```

In this example, `${name^^}` transforms the value of the name variable to uppercase before it's echoed, resulting in the output: `Hello, ALICE!`. You can also nest expansions, such as:

```
echo "Length of name: ${#name}"
```

Here, `${#name}` retrieves the length of the name variable, outputting: `Length of name: 5`. This showcases how powerful parameter expansion can be when used with commands like echo.

## Wildcards

Wildcards are special characters that can be used to match filenames or patterns.

**Common Wildcards:**

*   `*` (matches any number of characters, including none)
*   `?` (matches a single character)
*   `[...]` (matches any one of the enclosed characters)



````python
# @title Example: Lists all files ending in `.txt`.
%%bash
ls *.txt
````

````python
# @title Example: Lists files like `file1.txt`, `file2.txt`, but not `file10.txt`.
%%bash
touch file10.txt
ls file?.txt
````

````python
# @title Example: Lists `file1.txt`, `file2.txt`, and `file3.txt`.
%%bash
ls file[1-3].txt
````

Wildcards are special characters in bash that allow you to match patterns in filenames and directories, making it easier to perform operations on multiple files at once. For example, the asterisk `*` matches any number of characters, while the question mark `?` matches a single character.

If you have several text files in a directory, you could use `ls *.txt` to list all files that end with the `.txt` extension. Similarly, `rm file?.txt` would target files like `file1.txt` or `fileA.txt`, but not `file10.txt`.

Wildcards simplify file manipulation by enabling broader matching patterns, allowing users to work with groups of files efficiently. We will revisit this topic later as we explore more about paths and file management.
