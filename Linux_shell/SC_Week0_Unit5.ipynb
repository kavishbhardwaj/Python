# Week 0 - Unit 5: Networking Basics

## Introduction to Networking

Networking involves connecting computers and devices to share resources and communicate with each other. Understanding basic networking concepts is essential for troubleshooting and system management.




## 1. IP Address

### What is an IP Address?

An IP (Internet Protocol) address is a unique identifier assigned to each device on a network. It allows devices to communicate with each other over the internet or a local network.

### Types of IP Addresses:

- **IPv4**: 32-bit address, typically represented in decimal format (e.g., 192.168.1.1).
- **IPv6**: 128-bit address, designed to replace IPv4 (e.g., 2001:0db8:85a3:0000:0000:8a2e:0370:7334).

### Example Command to Check Your IP Address

In a terminal, you can use the following command to find your public IP address:

```bash
curl ifconfig.me


## 2. MAC Address

**What is a MAC Address?**
A MAC (Media Access Control) address is a hardware identifier assigned to network interfaces for communications on the physical network segment.

**Characteristics:**
* Format: Usually represented as six groups of two hexadecimal digits (e.g., 00:1A:2B:3C:4D:5E).
* Uniqueness: Designed to be unique to each network interface.

**Example Command to Check Your MAC Address**

You can find your MAC address with the following command:

```bash
iconfig
```


## 3. Domain Name

**What is a Domain Name?**

A domain name is a human-readable address that corresponds to an IP address. For example, `www.example.com` is a domain name that translates to an IP address.

**Structure of Domain Names:**

* Top-Level Domain (TLD): The last segment (e.g., `.com`, `.org`).
* Second-Level Domain: The part before the TLD (e.g., `example` in `www.example.com`).

## 4. Username

What is a Username?

A username is a unique identifier for a user on a system or network. It is used to log in and access resources. Usernames help in maintaining security and managing user permissions.

## 5. Common Networking Errors

**Wrong Domain Name**
If you enter an incorrect domain name, you'll likely encounter a "DNS lookup failed" error, which means the system couldn't resolve the domain to an IP address.

**Incorrect Username**
If the username is incorrect, you may receive an "authentication failed" error when trying to log in.

**Correct Username with Wrong IP Address or Domain Name**
If the username is correct but the IP address or domain name is wrong, you may see errors like "Connection refused" or "No route to host."

## 6. DNS Lookup

**What is DNS?**

DNS (Domain Name System) translates human-readable domain names into IP addresses. When you enter a URL in your browser, a DNS query is made to find the corresponding IP address.

**Performing a DNS Lookup**

You can perform a DNS lookup using the `nslookup` command:

```
nslookup www.example.com
```

## Conclusion

Understanding networking basics such as IP addresses, MAC addresses, domain names, usernames, and common errors is crucial for effective troubleshooting and network management.
